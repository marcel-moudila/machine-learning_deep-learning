```python
# pourcentage de valeurs manquantes par variable
# et lignes de valeurs manquantes par variable

colonnes = np.array(df.columns)
n = np.array(df.shape)[0]
for col in colonnes :
    index = []
    for i in range(n):
        if df[col].isnull()[i] == True:
            index = index + [i]
    print(col," : ", (len(index)/n)*100, " %"," , samples ",index,sep= "")
    print("_____________________________________________________________")
```
